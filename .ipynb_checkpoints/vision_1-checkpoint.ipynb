```python
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import img_as_ubyte
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.util import random_noise
import numpy as np
```

```python
image = imread('image1.jpg')

if image.ndim == 3:
    image = rgb2gray(image)

image = img_as_ubyte(image)

plt.imshow(image, cmap='gray')
plt.title('pre-processed image')
plt.axis('off')
plt.show()
```

```python
noisy_image = random_noise(image, mode='gaussian', var=0.1)
noisy_image = img_as_ubyte(noisy_image)

plt.imshow(noisy_image, cmap = 'gray')
plt.title('noisy image')
plt.axis('off')
plt.show()
```

```python
thresh = threshold_otsu (noisy_image)
binary_mask = noisy_image > thresh

plt.imshow(binary_mask, cmap = 'gray')
plt.title('Otsu Thresholding')
plt.axis('off')
plt.show()
```

```python
def region_growing(image, seed_point, tolerance=15):
    """
    Region growing from a single seed using 8-connectivity.
    """
    height, width = image.shape
    visited = np.zeros_like(image, dtype=bool)
    result = np.zeros_like(image, dtype=bool)

    seed_value = image[seed_point]
    stack = [seed_point]

    while stack:
        x, y = stack.pop()
        if visited[x, y]:
            continue

        visited[x, y] = True

        # Check if pixel is within tolerance
        if abs(int(image[x, y]) - int(seed_value)) <= tolerance:
            result[x, y] = True
            # Add 8-connected neighbors
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    if dx == 0 and dy == 0:
                        continue
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < height and 0 <= ny < width:
                        stack.append((nx, ny))
    return result
```

```python
seed = (0, 0)

grown_region = region_growing(noisy_image, seed, tolerance = 10)

plt.imshow(grown_region, cmap='gray')
plt.title('Region Grown from Seed')
plt.axis('off')
plt.show()
```
